13.1 Write the current date as a string to the text file today.txt

```python
from datetime import date
today = date.today()
with open('today.txt', 'w') as f:   
     f.write(str(today))

```

13.2 Read the text file today.txt into the string today_string.

```python
with open('today.txt', 'r') as f:
    today_string = f.read()
```

13.3 Parse the date from today_string.

```python
from datetime import datetime
format = '%Y-%m-%d'
parsed_date = datetime.strptime(today_string, format).date()
```

15.1 Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
import multiprocessing
import time
import random
from datetime import datetime

def show_time():
    wait_time = random.random()
    time.sleep(wait_time)
    print(f"process {multiprocessing.current_process().pid} waited {wait_time:.2f} seconds and finished at {datetime.now()}")

if __name__ == '__main__':
    processes = []
    for _ in range(3):
        p = multiprocessing.Process(target=show_time)
        processes.append(p)
        p.start()

    for p in processes:      
        p.join()
```
