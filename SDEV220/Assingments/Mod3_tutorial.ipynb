 Sort an Array of 0s, 1s, and 2s

```python
class Solution:
    def sort012(self, arr):
        low= 0 
        mid= 0
        high = len(arr) -1
        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low+= 1
                mid +=1
            elif arr[mid] ==1:
                mid+= 1
            else:
                arr[mid], arr[high]= arr[high], arr[mid]
                high -=1
```

 Binary Search

```python
class Solution:
    def binarysearch(self, arr, k):
        low, high = 0, len(arr) -1 
        answer = -1
        
        while low <= high:
            mid = (low + high)//2
            
            if arr[mid] ==k:
                answer = mid
                high= mid -1 
            elif arr[mid] < k:
                low = mid +1
            else:
                high = mid -1
        return answer
```
